```python
class Node:
    
    def __init__(self, data=None, nextt=None):
        self.data = data
        self.nextt = nextt
```

```python
class LinkedList:
    
    def __init__(self):
        self.head = None
    
    
    # Insert data at begining of linked list
    def insert_at_begining(self, data):
        node = Node(data, self.head)                   # Creates node object, from class Node ....
        self.head = node
    
    
    # Insert data at end of linked list
    def insert_at_end(self, data):
        if self.head is None:
            self.head = Node(data, None)
            return
        
        itr = self.head
        while itr.nextt:
            itr = itr.nextt
        itr.nextt = Node(data, None)
    
    
    # Create new linked list with list of items
    def insert_values(self, data_list):
        self.head = None
        for data in data_list:
            self.insert_at_end(data)
    
    
    # Print elements in the list
    def print_list(self):
        if self.head is None:
            print('Linked list is empty !!')
            return
        
        itr = self.head
        llstr = ''
        while itr:
            #print("Current ", itr, " --- Next ", itr.nextt)
            llstr += str(itr.data) + ' --> '
            itr = itr.nextt
            if itr is None:
                llstr += 'None'
        print(llstr)
        
        
    # Length of the linked list
    def get_length(self):
        count = 0
        itr = self.head
        while itr:
            count += 1
            itr = itr.nextt
        return count
    
    
    # Remove element at specific index
    def remove_at(self, index):
        
        if index < 0 or index >= self.get_length():
            raise Exception("Invalid index")    #print("Index out of range")
            return
        
        # Remove element at begining
        if index == 0:
            self.head = self.head.nextt
            return
        
        # Remove element elsewhere
        count = 0
        itr = self.head
        while itr:
            if count == index-1:
                itr.nextt = itr.nextt.nextt
                break
            count += 1
            itr = itr.nextt
            
            
    # Insert element at specific index
    def insert_at(self, index, data):
        if index < 0 or index > self.get_length():
            raise Exception("Invalid index")
            return
        
        # Insert at begining
        if index == 0:
            self.insert_at_begining(data)
            return
        
        # Other cases
        count = 0
        itr = self.head
        while itr:
            if count == index-1:
                node = Node(data, itr.nextt)
                itr.nextt = node
                break
                
            count += 1
            itr = itr.nextt
            
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
        
class LinkedList:
    def __init__(self):
        self.head = None
    
    def printList(self):
        last_node = self.head
        while last_node:
            print(last_node.data, end=' ')
            last_node = last_node.next
        print()
    
    
    # INSERTION
    # APPEND
    def append(self, data):
        newNode = Node(data)
        
        if self.head is None:
            self.head = newNode
            return
        
        last_node = self.head
        while last_node.next:
            last_node = last_node.next
            
        last_node.next = newNode
        
    # PREPEND
    def prepend(self, data):
        newNode = Node(data)
        
        newNode.next = self.head
        self.head = newNode
    
    # MIDDLE
    def insert(self, prev_node, data):
        newNode = Node(data)
        
        if not prev_node:
            print('Given node not in list')
            return
        
        newNode.next = prev_node.next
        prev_node.next = newNode   
    
    
    # DELETION
    # DELETE NODE WITH SPECIFIC VALUE
    def delete_node(self, key):
        cur_node = self.head
        
        # Head node
        if cur_node and cur_node.data == key:
            self.head = cur_node.next
            cur_node = None
            return 
    
        # Other node
        prev = None
        while cur_node and cur_node.data != key:
            prev = cur_node
            cur_node = cur_node.next
        
        if cur_node is None:
            return
        prev.next = cur_node.next    
        cur_node = None
    
    # DELETE AT POS
    def delete_node_at_pos(self, pos):
        cur_node = self.head
        
        if pos == 0:
            self.head = cur_node.next
            cur_node = None
            return
        
        count = 0
        while cur_node and pos != count:
            prev = cur_node
            cur_node = cur_node.next
            count += 1
        prev.next = cur_node.next
        
        
    # REVERSE
    def reverse_iterative(self):
        prev = None
        cur = self.head
        while cur:
            nxt = cur.next
            cur.next = prev
            prev = cur
            cur = nxt
        self.head = prev
            
    def reverse_recursive(self):
        def _reverse_recursive(cur, prev):
            if not cur:
                return prev
            nxt = cur.next
            cur.next = prev
            prev = cur
            cur = nxt
            return _reverse_recursive(cur, prev)
            
        self.head = _reverse_recursive(cur=self.head, prev=None) 
    
    
    # LENGTH OF THE LIST
    def len_iterative(self):
        count = 0
        curr_node = self.head
        while curr_node:
            count += 1
            curr_node = curr_node.next
        return count
    
    def len_recursive(self, node):
        if node is None:
            return 0
        return 1 + self.len_recursive(node.next)
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
         
        
class CircularLinkedList:
    def __init__(self):
        self.head = None
    
    def printList(self):
        cur = self.head
        while cur:
            print(cur.data, end=' ')
            cur = cur.next
            if cur == self.head:
                break
        print()
    
    # APPEND
    def append(self, data):
        if not self.head:
            self.head = Node(data)
            self.head.next = self.head
        else:
            newNode = Node(data)
            cur = self.head
            while cur.next != self.head:
                cur = cur.next
            cur.next = newNode
            newNode.next = self.head      
                
    # PREPEND
    def prepend(self, data):
        if not self.head:
            self.head = Node(data)
            self.head.next = self.head
        else:
            newNode = Node(data)
            cur = self.head
            while cur.next != self.head:
                cur = cur.next
            cur.next = newNode
            newNode.next = self.head
            self.head = newNode
                
    # REMOVE NODE
    def remove(self, key):
        cur = self.head
        if self.head.data == key:
            while cur.next != self.head:
                cur = cur.next
            cur.next = self.head.next
            self.head = self.head.next
        else:
            prev = None
            while cur.next != self.head:
                prev = cur
                cur = cur.next
                if cur.data == key:
                    prev.next = cur.next
                    cur = cur.next
                    
    # JOSEPHUS PROBLEM
    # Remove node
    def remove_node(self, node):
        cur = self.head
        if self.head == node:
            while cur.next != self.head:
                cur = cur.next
            cur.next = self.head.next
            self.head = self.head.next
        else:
            prev = None
            while cur.next != self.head:
                prev = cur
                cur = cur.next
                if cur == node:
                    prev.next = cur.next
                    cur = cur.next
                    
    def josephus_circle(self, step):
        cur = self.head
        prev = None
        while cur != cur.next:    # or len(self) > 1
            count = 1
            while count != step:
                cur = cur.next
                count += 1
            self.remove_node(cur)
            # print(cur.data)  # Removed data
            cur = cur.next
```
